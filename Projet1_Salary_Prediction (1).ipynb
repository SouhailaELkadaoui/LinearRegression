## <b>Projet<b>:<i>   La modélisation de salaire   en fonction des années d'experiences<i>

Le modéle utilisé est la régression linéaire simple

La régression linéaire est  un modèle parmi plusieurs de machine learning.Il s'appuie sur la relation linéaire existante entre les attributes et  le variable  cible. Pour introduire cette relation à un modèle linéaire  qui nous permet de trouver les valeurs de variable cible, on  se base dans la construction de modèle sur l' historique   pour estimer les meilleurs paramètres optimales de la fonction modélisée.

#### Description de Dataset

La dataset est une ensemble des exemples des mantants   par rapport à l'expérience de salarie

#### Importation  des librairies nécessaires

```python
import pandas as pd
import numpy as np
from matplotlib  import pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

#### Importation de Dataset

```python
dataset=pd.read_csv('Salary_Data.csv',header=0)
#L'affichage des 1 emes lignes de dataset


dataset.size
```

#### L'exploration de Dataset

```python
#La vérification de l'existance des valeurs manquantes
dataset.isnull().sum()
```

YearExperience     :   0<br>
Salary            :   0<br>


```python
#La visualisation de data set
Attributs=dataset['YearsExperience']
Target=dataset['Salary']


#plot:
plt.figure(figsize=(20,10))
plt.scatter(Attributs,Target,label='Real')
plt.xlabel('YearsExperience')
plt.ylabel(' Salary' )
plt.title('La variation de la salaire  en fonction des années des experiences')
```

```python
#L'etude de l'existence des valeurs aberrants
plt.figure(figsize=(15,8))
plt.boxplot(dataset,labels=dataset.columns)
            
```

#### Preparation des données

```python
# Split le dataset en 2 parties:train et test
#random_state. Ce paramètre est utilisé pour initialiser le générateur de nombres pseudo-aléatoires qui est utilisé pour diviser les données de manière aléatoire.
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]

X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.8, random_state=2529)
```

```python
# Commence de Modélisation
model=linear_model.LinearRegression()
#trainnig
model.fit(X_train,y_train)
```

```python
# les parametres  de model
model.intercept_
```

```python
model.coef_
```

```python
# la prédiction des prochaines valeurs de partie test
y_pred = model.predict(X_test)


```

```python
#La visualisation
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color='r')
```

```python
# Calculer la performance de modèle
R = r2_score(y_test, y_pred)
print("La performance du modèle est : {}".format(R))
```
